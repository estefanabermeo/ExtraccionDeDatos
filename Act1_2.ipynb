```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Datos de Facturacion.xlsx')
df
```

```python
# Filtro a
filtroa =df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtroa
```

```python
filtroa.to_csv("filtroa.csv")
```

```python
# Filtro b
filtrob = df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]
filtrob
```

```python
filtrob.to_csv("filtrob.csv")
```

```python
# Filtro c
filtroc = df[df["FECHA_ENT"] == "2022-02-28 00:00:00"]
filtroc
```

```python
filtroc.to_csv("filtroc.csv")
```

```python
# Filtro d
filtrod =df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtrod
```

```python
filtrod.to_csv("filtrod.csv")
```

```python
# Filtro e
filtroe = df.iloc[ : , [0,6,7,9]]  #Columnas no consecutivas 
filtroe
```

```python
filtroe.to_csv("filtroe.csv")
```

```python
# Filtro f
filtrof = df.iloc[7001:7100,: ]  
filtrof
```

```python
filtrof.to_csv("filtrof.csv")
```
