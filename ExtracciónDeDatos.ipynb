1. Filtro por objeto
Las variables filtradas fueron la columna "NOMBRE_VENEDEDOR" pero únicamente las que correspondan a "LETICIA RAMIREZ HERNANDEZ"

```python
# 1. Filtro por objeto
# Las variables filtradas fueron la columna "NOMBRE_VENEDEDOR" pero únicamente las que correspondan a "LETICIA RAMIREZ HERNANDEZ"
filtro1=df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1

# Convertir archivo filtrado a CSV
filtro1.to_csv("filtro1.csv")
```

```python
# 2. Filtro por filas
# Las variables filtradas fueron todas la columnas pero únicamente las filas 1, 25 y 50
#filtro2= df.iloc[2:15,: ]           #Filas consecutivas
filtro2=df.iloc[ [1, 25, 50] , : ]  #Filas no consecutivas 
filtro2

# Convertir archivo filtrado a CSV
filtro2.to_csv("filtro2.csv")
```

```python
#3. Filtro por columnas
# Las variables filtradas fueron únicamente las columnas consecutivas de la número 1 a la 5 y todos sus registros
filtro3=df.iloc[ : , 1:5]       #Columnas consecutivas 
#filtro3=df.iloc[ : , [3,5,12]]  #Columnas no consecutivas 
filtro3

# Convertir archivo filtrado a CSV
filtro3.to_csv("filtro3.csv")
```

```python
#4. Filtro de filas vs columnas 
#Paso 1: Leemos el contenido del archivo
#from google.colab import files
#uploaded=files.upload()
```

```python
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df1
```

```python
df1 = df1.set_index('CVE_ART')
df1
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas 
# Las variables filtradas fueron el ínidce CVE_ART (sólo aquellos registros con valor "TUE00244" y "RB00007") y el PRECIO_UNITARIO
filtro4=df1.loc[["TUE00244","RB00007"], ["PRECIO_UNITARIO"]]
filtro4

# Convertir archivo filtrado a CSV
filtro4.to_csv("filtro4.csv")
```

```python
# 5. Filtro por cabecera
# El filtro se aplicó a los primeros 5 registros de todas las columnas/variables
filtro5 = df.head(5)
filtro5

# Convertir archivo filtrado a CSV
filtro5.to_csv("filtro5.csv")
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
# 6. Filtro por comparacion
# El filtro se aplicó a la variable "PRECIO_UNITARIO" para aquellos registros con valor superior a 10,000
filtro6 = df[df["PRECIO_UNITARIO"] > 10000]
filtro6

# Convertir archivo filtrado a CSV
filtro6.to_csv("filtro6.csv")
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
# 7. Filtro Y
# El filtro se aplicó a las variables PRECIO_UNITARIO menor a 8,000 y a la FECHA_DOC correspondiente al 5 de enero 2022
filtro7=df[(df["PRECIO_UNITARIO"] < 8000) & (df["FECHA_DOC"] == "2022-01-05")]
filtro7

# Convertir archivo filtrado a CSV
filtro7.to_csv("filtro7.csv")
```

```python
# 8. Filtro o
# El filtro se aplicó a las variables PRECIO_UNITARIO menor a 8,000 ó a la FECHA_DOC correspondiente al 5 de enero 2022
filtro8=df[(df["PRECIO_UNITARIO"] < 8000) | (df["FECHA_DOC"] == "2022-01-05")]
filtro8

# Convertir archivo filtrado a CSV
filtro8.to_csv("filtro8.csv")
```

```python
# 9. Filtro not
# El filtro se aplicó a todas las variables excepto al PRECIO_UNITARIO mayor o igual a 8,000 y tampoco a la FECHA_DOC correspondiente al 5 de enero 2022
filtro9=df[~(df["PRECIO_UNITARIO"] >= 8000) & ~(df["FECHA_DOC"] == "2022-01-05")]
filtro9

# Convertir archivo filtrado a CSV
filtro9.to_csv("filtro9.csv")
```
